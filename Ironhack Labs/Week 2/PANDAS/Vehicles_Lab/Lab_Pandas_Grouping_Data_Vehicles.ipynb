Converting Grams/Mile to Grams/Km

1 Mile = 1.60934 Km

Grams/Mile * Mile/Km -> Grams/Mile * 1 Mile/1.60934Km

$$ \frac{Grams}{Mile} * \frac{Mile}{Km} $$

$$ \frac{Grams}{Mile} * \frac{1 Mile}{1.60934Km}  $$

convert MPG columns to km_per_liter

MPG = Miles/Gallon -> Km/Liter

1 Mile = 1.60934 Km

1 Gallon = 3.78541 Liters

$$ \frac{Miles}{Gallon} -> \frac{Miles}{Gallon} * \frac{Km}{Miles} * \frac{Gallon}{Liters}$$

$$ \frac{Miles}{Gallon} -> \frac{Miles}{Gallon} * \frac{1.60934Km}{ 1Miles} * \frac{1 Gallon}{3.78541 Liters}$$

* ( 1.60934 / 3.78541 )


What brand has the most cars?

```python
groupby_df = (cars_df.groupby(['Make'])['Model'].agg(['count']).sort_values(by='count', ascending=False).reset_index().drop_duplicates('Make', keep='first'))
groupby_df
```

<b>show the average CO2_Emission_Grams/Km  by Brand

```python
### your code us here
cars_df["CO2 Emission Grams/Mile"] = cars_df["CO2 Emission Grams/Mile"]  /1.60934
cars_df.rename(columns = {"CO2 Emission Grams/Mile":"CO2 Emission Grams/Km"}, inplace = True)
cars_df
```

```python
#cars_df.groupby(["CO2 Emission Grams/Km"]).agg({"CO2 Emission Grams/Km":{"CO2 Emission Grams/Km AVG":"mean"}})
```

```python
avg_CO2 = cars_df.groupby("Make").agg({"CO2 Emission Grams/Km":"mean"})

cars_df["CO2 Emission Grams/Km AVG"] = avg_CO2

avg_CO2
```

<b>show the average CO2_Emission_Grams/Km  by Brand ... sorted

```python
### your code us here
sorted_val  = avg_CO2.sort_values("CO2 Emission Grams/Km", ascending = False)
sorted_val
```

(Optional) Use `pd.cut` or `pd.qcut` to create 4 groups (bins) of cars, by Year. We want to explore how cars have evolved decade by decade.

```python
cars_df['Year'].describe()
```

```python
#We want tto know how cars have evolved decade by decade
year_range = pd.cut(cars_df['Year'], bins=4, precision = 0).value_counts(sort=False)
year_range
```

```python
cars_df['year_range'] = pd.cut(x=cars_df['Year'], bins=[1980,1990,2000,2010,2020], labels=['80s','90s','20s','21s'])
cars_df


cars_df.groupby("year_range").agg({"Year":"count"})
```
