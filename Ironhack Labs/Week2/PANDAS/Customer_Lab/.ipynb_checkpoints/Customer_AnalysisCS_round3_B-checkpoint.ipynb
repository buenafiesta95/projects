```python
# create first figure and ax
fig, ax = plt.subplots(4,2,figsize = (12,10))

# we add a title to the figure or the ax
fig.suptitle('Distribution plots for numerical values', fontsize=18)
# we plot the different lines adding them to the axis
sns.displot(df_num.customer_lifetime_value, ax= ax[0,0], label='customer_lifetime_value')
sns.displot(df_num.income, ax= ax[1,0], c='red', label='income')
sns.displot(df_num.monthly_premium_auto, ax=ax[2,0], c='green', label='monthly_premium_auto')
sns.displot(df_num.months_since_last_claim, ax=ax[3,0], c='orange', label='months_since_last_claim')
sns.displot(df_num.months_since_policy_inception, ax=ax[0,1], c='orange', label='months_since_policy_inception')
sns.displot(df_num.number_of_open_complaints, ax=ax[1,1], c='orange', label='number_of_open_complaints')
sns.displot(df_num.number_of_policies, ax=ax[2,1], c='orange', label='number_of_policies')
sns.displot(df_num.total_claim_amount,ax=ax[3,1], c='orange', label='total_claim_amount')
```
