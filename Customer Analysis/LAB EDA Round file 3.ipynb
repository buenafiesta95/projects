```python
def lower_case_column_names(ca_df):
    ca_df.columns=[i.lower() for i in ca_df.columns]
    return ca_df
```

```python
lower_case_column_names(ca_df)
```

```python
def rename_columns(ca_df):
    ca_df.rename(columns={'customer':'id','st':'state', 'gender':'gender',"education":"education", "customer lifetime value":"CLV","monthly premium auto": "monthly premium auto","policy type":"policy type", "vehicle class" : "vehicle class", "total claim amount":"total claim amount"}, inplace=True )
    return ca_df
```

```python
rename_columns(ca_df)
```

```python
def drop_columns(ca_df) :
    ca_df.drop(columns=["id"], inplace=True)
    return ca_df
```

```python
drop_columns(ca_df)
```
